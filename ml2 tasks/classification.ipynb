```python
import pandas as pd, matplotlib.pyplot as plt, numpy as np, seaborn as sb
```

```python
df1 = pd.read_csv('./data/customer segmentation/Test.csv')
df2 = pd.read_csv('./data/customer segmentation/Train.csv')
```

```python
frames = [df1, df2]
customers_data = pd.concat(frames)
```

```python
customers_data
```

```python
customers_data.dtypes
```

```python
customers_data['Gender'] = customers_data['Gender'].map({'Male': 0,'Female': 1})
customers_data['Ever_Married'] = customers_data['Ever_Married'].map({'No': 0,'Yes': 1})
customers_data['Graduated'] = customers_data['Graduated'].map({'No': 0,'Yes': 1})
customers_data['Spending_Score'] = customers_data['Spending_Score'].map({'Low': 0,'Average': 1,'High': 2})
customers_data['Segmentation'] = customers_data['Segmentation'].map({'A': 0,'B': 1,'C': 2, 'D': 3})

print(customers_data['Segmentation'].unique())
print(customers_data.isna().sum())
```

```python
ax3 = sb.violinplot(x='Age', data=customers_data)
```

```python
ax4 = sb.violinplot(x='Spending_Score', data=customers_data)
```

```python
ax3 = sb.violinplot(x='Segmentation', data=customers_data)
```

```python
y = customers_data['Segmentation']
X = customers_data.drop('Segmentation', axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                   random_state=104, 
                                   test_size=0.30, 
                                   shuffle=True)

X_val, X_test, y_val, y_test = train_test_split(X_test, y_test,
                                   random_state=104, 
                                   test_size=0.50, 
                                   shuffle=True)
```
